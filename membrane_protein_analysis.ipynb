```python
# @title Display the data
df = pd.read_csv("membrane_protein_analysis_final.csv")
df.head()

from itables import show

show(df,
     classes="compact",
     buttons=["columnsToggle"],
     layout={"top1": "searchBuilder"},
     searchBuilder={
        "preDefined": {
            "criteria": [
                {"data": "CtermDis", "condition": "<", "value": ["50"]}
            ]
        }
    },
)
```

```python
gene_list = df.loc[df['cterm_distance'] <= 100, 'Gene Names'].tolist()
print(gene_list)

gene_list_series = pd.Series(gene_list)
gene_list_series.to_csv('gene_list_cterm_100.csv', index=False)
```

```python
all_gene_list = df['Gene Names']
print(all_gene_list)

all_gene_list.to_csv('all_gene_list.csv', index=False)
```

```python
import plotly.express as px

fig1 = px.histogram(df, x="cterm_distance", color="in_biogrid", marginal="rug")
fig2 = px.histogram(df, x="cterm_distance", color="in_massspec", marginal="rug")
fig3 = px.histogram(df, x="cterm_distance", color="N_term_md", marginal="rug")
fig1.show()
fig2.show()
fig3.show()
```
